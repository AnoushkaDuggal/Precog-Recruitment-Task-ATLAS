```python
import networkx as nx
import plotly.graph_objects as go

# Extract country names as a list from the DataFrame
country_names = countries['Name'].tolist()  # This gets the actual country names

# Build directed graph by last/first letter
G = nx.DiGraph()
G.add_nodes_from(country_names)  # Use country_names instead of countries
for a in country_names:          # Use country_names instead of countries
    for b in country_names:      # Use country_names instead of countries
        if a != b and a.strip()[-1].lower() == b.strip()[0].lower():
            G.add_edge(a, b)

# Layout for nodes
pos = nx.spring_layout(G, k=0.15, iterations=50)
x_nodes = [pos[k][0] for k in G.nodes()]
y_nodes = [pos[k][1] for k in G.nodes()]

# Edge coordinates
edge_x = []
edge_y = []
for e in G.edges():
    x0, y0 = pos[e[0]]
    x1, y1 = pos[e[1]]
    edge_x += [x0, x1, None]
    edge_y += [y0, y1, None]

# Make plotly graph
edge_trace = go.Scatter(x=edge_x, y=edge_y,
                        line=dict(width=0.5, color='#888'),
                        hoverinfo='none', mode='lines')

node_trace = go.Scatter(x=x_nodes, y=y_nodes,
                        mode='markers+text',
                        text=list(G.nodes()),
                        hoverinfo='text',
                        marker=dict(showscale=True,
                                    color='skyblue',
                                    size=10,
                                    line=dict(width=2)),
                        textposition="top center")

fig = go.Figure(data=[edge_trace, node_trace],
                layout=go.Layout(showlegend=False,
                                 hovermode='closest',
                                 margin=dict(b=0,l=0,r=0,t=0),
                                 xaxis=dict(showgrid=False, zeroline=False),
                                 yaxis=dict(showgrid=False, zeroline=False),
                                 title="Interactive Country Chain Graph"))

fig.show()
```

```python
import pandas as pd
import networkx as nx
import plotly.graph_objects as go
from unidecode import unidecode


country_names = countries['Name'].tolist()  # Convert to list of strings

# Create directed graph
G = nx.DiGraph()
G.add_nodes_from(country_names)                 # Use country_names instead of countries
for a in country_names:                          # Use country_names instead of countries
    for b in country_names:
        if a != b and a.strip()[-1].lower() == b.strip()[0].lower():
            G.add_edge(a, b)

# Layout positions
pos = nx.spring_layout(G, k=0.15, iterations=50)
node_x = [pos[node][0] for node in G.nodes()]
node_y = [pos[node][1] for node in G.nodes()]

# Build edge coordinates for Plotly
edge_x, edge_y = [], []
for edge in G.edges():
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    edge_x += [x0, x1, None]
    edge_y += [y0, y1, None]

edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=0.5, color='#888'),
    hoverinfo='none', mode='lines'
)

# Prepare hover text with in-degree and out-degree info
hover_texts = []
for node in G.nodes():
    indeg = G.in_degree(node)
    outdeg = G.out_degree(node)
    text = f"<b>{node}</b><br>Pointed to by: {indeg}<br>Points to: {outdeg}"
    hover_texts.append(text)

node_trace = go.Scatter(
    x=node_x, y=node_y,
    mode='markers+text',
    text=[node for node in G.nodes()],
    hoverinfo='text',
    hovertext=hover_texts,
    marker=dict(
        size=10,
        color='skyblue',
        line_width=2
    ),
    textposition="top center"
)

fig = go.Figure(data=[edge_trace, node_trace],
    layout=go.Layout(
        showlegend=False,
        hovermode='closest',
        margin=dict(b=0,l=0,r=0,t=30),
        xaxis=dict(showgrid=False, zeroline=False),
        yaxis=dict(showgrid=False, zeroline=False),
        title="Interactive Country Chain Directed Graph"
    ))

fig.show()
```

```python
import pandas as pd
import networkx as nx
from unidecode import unidecode
import collections
import plotly.graph_objects as go

# --- Data Loading and Graph Construction ---
# We will use the same reliable data source and graph as before.
url = "https://datahub.io/core/country-list/r/data.csv"
df = pd.read_csv(url)
countries = df['Name'].apply(unidecode).tolist()

G = nx.DiGraph()
G.add_nodes_from(countries)
for a in countries:
    for b in countries:
        if a != b and len(a.strip()) > 0 and len(b.strip()) > 0:
            if a.strip()[-1].lower() == b.strip()[0].lower():
                G.add_edge(a, b)

# --- Probabilistic Analysis from 'S' Countries ---

# Step 1: Find all countries that start with 'S'
s_countries = [country for country in G.nodes() if country.strip().upper().startswith('S')]

# Step 2: Get all the last letters of these 'S' countries
ending_letters = [country.strip()[-1].upper() for country in s_countries]

# Step 3: Count the occurrences of each ending letter
letter_counts = collections.Counter(ending_letters)

# Step 4: Calculate the probability of each ending letter
total_s_countries = len(s_countries)
letter_probabilities = {letter: count / total_s_countries for letter, count in letter_counts.items()}

# Create a DataFrame for easy plotting
prob_df = pd.DataFrame(letter_probabilities.items(), columns=['Ending Letter', 'Probability'])
prob_df = prob_df.sort_values(by='Probability', ascending=False)

print(prob_df)

# --- Step 5: Visualize the Probabilities with a Bar Chart ---
fig = go.Figure(go.Bar(
    x=prob_df['Ending Letter'],
    y=prob_df['Probability'],
    marker=dict(color=prob_df['Probability'], colorscale='YlGnBu'),
))

fig.update_layout(
    title="Probability of Opponent's Next Starting Letter from an 'S' Country",
    xaxis_title="Opponent's Next Starting Letter",
    yaxis_title="Probability",
)

fig.show()
```

```python
import pandas as pd
import networkx as nx
from unidecode import unidecode
import re
import collections
import plotly.graph_objects as go

# --- Data Loading and Cleaning ---
url = "https://datahub.io/core/country-list/r/data.csv"
df = pd.read_csv(url)

country_names = df['Name'].apply(unidecode).tolist()
country_names = [re.sub(r'\s*\(.*?\)\s*$', '', name).strip() for name in country_names]
country_names = sorted(list(set(country_names)))

# --- Atlas Graph Construction ---
G = nx.DiGraph()
G.add_nodes_from(country_names)

for a in country_names:
    for b in country_names:
        if a != b and len(a.strip()) > 0 and len(b.strip()) > 0:
            if a.strip()[-1].lower() == b.strip()[0].lower():
                G.add_edge(a, b)

print("Atlas game graph (Country Only) created successfully.")


# --- Refined Strategy: Find the Best Move to Minimize Opponent's Options ---

# Step 1: Identify the strategic value of each last letter
# This is the count of countries that start with that letter.
first_letter_counts = collections.Counter(country[0].upper() for country in G.nodes() if len(country) > 0)

# Step 2: Choose a starting country
start_country = 'Serbia'  # You can change this to any country starting with 'S'

if start_country not in G.nodes():
    print(f"\nError: Starting country '{start_country}' is not in the graph. Please choose a valid country.")
else:
    print(f"\nAnalyzing optimal moves from '{start_country}'...")

    # Find all possible moves from the starting country
    possible_moves = list(G.successors(start_country))

    if not possible_moves:
        print(f"No moves possible from '{start_country}'.")
    else:
        # Step 3: Evaluate each possible move
        move_evaluations = []
        for move in possible_moves:
            last_letter_of_move = move.strip()[-1].upper()

            # The strategic value is the number of options the opponent will have
            strategic_value = first_letter_counts.get(last_letter_of_move, 0)

            move_evaluations.append({
                'country': move,
                'last_letter': last_letter_of_move,
                'opponent_options': strategic_value
            })

        # Sort moves to find the one that leaves the opponent with the fewest options
        best_move_df = pd.DataFrame(move_evaluations).sort_values(by='opponent_options', ascending=True)

        # --- Output the Analysis ---
        print("\nPossible moves ranked by opponent's remaining options:")
        print(best_move_df)

        # Highlight the optimal move
        optimal_move = best_move_df.iloc[0]['country']
        optimal_last_letter = best_move_df.iloc[0]['last_letter']
        optimal_options = best_move_df.iloc[0]['opponent_options']

        print(f"\n--- Optimal Strategic Move from '{start_country}' ---")
        print(f"Move to: {optimal_move}")
        print(f"This leaves your opponent with only {optimal_options} possible move(s) starting with '{optimal_last_letter}'.")
```

```python
import pandas as pd
import networkx as nx
from unidecode import unidecode
import re
import collections
import plotly.graph_objects as go

# --- Data Loading and Cleaning ---
url = "https://datahub.io/core/country-list/r/data.csv"
df = pd.read_csv(url)

country_names = df['Name'].apply(unidecode).tolist()
country_names = [re.sub(r'\s*\(.*?\)\s*$', '', name).strip() for name in country_names]
country_names = sorted(list(set(country_names)))

# --- Atlas Graph Construction ---
G = nx.DiGraph()
G.add_nodes_from(country_names)

for a in country_names:
    for b in country_names:
        if a != b and len(a.strip()) > 0 and len(b.strip()) > 0:
            if a.strip()[-1].lower() == b.strip()[0].lower():
                G.add_edge(a, b)

print("Atlas game graph (Country Only) created successfully.")


# --- Corrected Strategy: Find the Optimal Starting Country and First Move ---

# Step 1: Count how many countries start with each letter
first_letter_counts = collections.Counter(country[0].upper() for country in G.nodes() if len(country) > 0)
print("\nStrategic value of each letter (number of countries that start with it):")
print(first_letter_counts)


# Step 2: Find all countries that start with 'S'
s_countries = [country for country in G.nodes() if country.strip().upper().startswith('S')]

# Step 3: Evaluate each 'S' country as a potential first move
move_evaluations = []
for country in s_countries:
    last_letter_of_move = country.strip()[-1].upper()

    # The strategic value of this move is the number of options the opponent will have
    opponent_options = first_letter_counts.get(last_letter_of_move, 0)

    move_evaluations.append({
        'country': country,
        'last_letter': last_letter_of_move,
        'opponent_options': opponent_options
    })

# Step 4: Find the move that leaves the opponent with the fewest options
best_move_df = pd.DataFrame(move_evaluations).sort_values(by='opponent_options', ascending=True)

# --- Output the Analysis ---
print(f"\nEvaluating all possible starting countries that begin with 'S'...")
print("\nRanked starting countries by how few options they leave the opponent:")
print(best_move_df)

optimal_start_country = best_move_df.iloc[0]['country']
optimal_last_letter = best_move_df.iloc[0]['last_letter']
optimal_options = best_move_df.iloc[0]['opponent_options']

print(f"\n--- Optimal First Move ---")
print(f"To maximize your advantage, you should start with the country: {optimal_start_country}")
print(f"This move ends with the letter '{optimal_last_letter}', leaving your opponent with only {optimal_options} possible move(s).")
```
